```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
prosper_data = pd.read_csv('prosperLoanData.csv')
```

```python
prosper_data.head()
```

```python
prosper_data.info()
```

**EMPLOYMENT STATUS**

Observing how the number of loans are given depending on the 'Employment Status'.

```python
base_color = sb.color_palette()[0]
cat_order = prosper_data['EmploymentStatus'].value_counts().index
sb.countplot(data=prosper_data, x='EmploymentStatus', color=base_color, order=cat_order)
plt.title('Loans Vs Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=90);
```

**INCOME RANGE**

Understanind the distribution of income range for people applying for loans.

```python
base_color = sb.color_palette()[0]
cat_order = prosper_data['IncomeRange'].value_counts().index
sb.countplot(data=prosper_data, x='IncomeRange', color=base_color, order=cat_order)
plt.title('Income range of applicants')
plt.ylabel('Count')
plt.xticks(rotation=90);
```

It looks like the majority of applicants have an income between 25K and 75K.

**LOAN STATUS**

We want to understand how the loans are distributed in the various categories showing either good standing or some sort of deliquency. 

```python
# First we want to order the loans from best status to worst status 
status_loans = prosper_data['LoanStatus'].value_counts()
status_loans
```

```python
loan_status_order = ['Completed', 'FinalPaymentInProgress', 'Current', 'Past Due (1-15 days)',
                    'Past Due (16-30 days)', 'Past Due (31-60 days)', 'Past Due (61-90 days)',
                    'Past Due (91-120 days)', 'Past Due (>120 days)', 'Defaulted', 'Chargedoff',
                    'Cancelled']
```

```python
base_color = sb.color_palette()[0]
plt.figure(figsize=[10, 6])

sb.countplot(data = prosper_data, x = 'LoanStatus', color = base_color,
             order = loan_status_order)

for i in range(len(loan_status_order)):
    count = status_loans[loan_status_order[i]]
    #print(ls_order[i], ': ', count)
    percentages = '{:.1f}%'.format(100 * count / sum(status_loans))
    plt.text(i, count + 1, percentages, va = 'bottom', ha = 'center' )

plt.title('Distribution of Loan Status')
plt.xticks(rotation = 90)
plt.yticks(np.arange(0, 70000, 5000))

plt.xlabel('Loan Status')
plt.ylabel('Count)');
```

**BORROWER APR**

```python
bin_edges = np.arange(0, prosper_data['BorrowerAPR'].max()+0.05, 0.02)
plt.hist(data = prosper_data, x = 'BorrowerAPR', bins=bin_edges)
plt.title('Borrower APR Distribution')
plt.xlabel('Borrower APR')
plt.ylabel('Count');
```

```python
bin_edges = np.arange(0, prosper_data['BorrowerRate'].max()+0.05, 0.02)
plt.hist(data = prosper_data, x = 'BorrowerRate', bins=bin_edges)
plt.title('Borrower Rate Distribution')
plt.xlabel('Borrower Rate')
plt.ylabel('Count');
```

**Borrower APR and Rate** seem to follow a very similar distribution, unimodal slightly left-skewed.

**MONTHLY PAYMENT**

```python
bin_edges = np.arange(0, prosper_data['MonthlyLoanPayment'].max()+10, 50)
plt.hist(data = prosper_data, x = 'MonthlyLoanPayment', bins=bin_edges)
plt.yscale('log');
```

It looks like the majority of monthly payments is below 1000 dollars and only a low amount is more than 2000 dollars a month.

**CREDIT SCORE**

```python
prosper_data['CreditRange'] = prosper_data['CreditScoreRangeLower'].astype(str).str.extract(r'(\d{1,3})') + ' - ' + prosper_data['CreditScoreRangeUpper'].astype(str).str.extract(r'(\d{1,3})')
```

```python
credit_order = list(prosper_data['CreditRange'].value_counts().sort_index().index)

#credit_order = credit_order[-1:] + credit_order[:-1]
```

```python
cr_classes = pd.api.types.CategoricalDtype(ordered = True, categories = credit_order)

prosper_data['CreditRange'] = prosper_data['CreditRange'].astype(cr_classes)
```

```python
type_counts = prosper_data['CreditRange'].value_counts()
```

```python
fig = plt.figure(figsize = (15, 10))

sb.countplot(data = prosper_data, x = 'CreditRange', color = base_color,
                    order = credit_order)

for i in range(len(type_counts)):
    count = type_counts[i]
    #print(count)
    display_val = '{:.1f}%'.format(100 * count / sum(type_counts))
    plt.text(i, count + 1, display_val, va = 'bottom', ha = 'center')

plt.xticks(rotation = 90)
plt.xlabel('Credit Range')
plt.ylabel('Frequency')

plt.title('Credit Range of Borrowers');
```
